````python
from ddgs import DDGS

topic = input("Enter Topic: ")

all_content = []

with DDGS() as ddgs:

    search_results = ddgs.text(
        topic,
        max_results=25
    )

    
````

````python
for result in search_results:

        url = result["href"]

        print(f"\nExtracting: {url}")

        try:
# extract content from the URL in markdown format
            article = ddgs.extract( 
                url,
                fmt="text_markdown"
            )

            content = article["content"]

            all_content.append(content)

            print(content[:500])

        except Exception as e:

            print("Error:", e)
````

````python
import os
import json
from dotenv import load_dotenv
from langchain_groq import ChatGroq


load_dotenv()

api_key = os.getenv("API_KEY")

if not api_key:
    raise ValueError("API_KEY not found in .env")


client = ChatGroq(
    groq_api_key=api_key,
    model_name="llama-3.3-70b-versatile",
    temperature=0.3,
)


# Storing summaries for each source
Final_summaries = []

# Loop through each content and generate summary
for idx, content in enumerate(all_content):
    # from the above command it means idx is the index of the content in the all_content list and content is the actual content extracted from the URL.

    print(f"\nProcessing Source {idx+1}")

    # Prevent huge inputs
    content = content[:5000]

    prompt = f"""
You are an expert research analyst.

Analyze the following content and return ONLY valid JSON.

Required JSON Structure:

{{
    "summary": "150 word summary of the content",
}}

Topic:
{topic}

Content:
{content}
"""

    try:

        response = client.invoke(prompt)

        cleaned_response = response.content.strip()

        # Remove markdown if present
        cleaned_response = cleaned_response.replace("```json", "")
        cleaned_response = cleaned_response.replace("```", "")
        cleaned_response = cleaned_response.strip()

        parsed_json = json.loads(cleaned_response)

        Final_summaries.append({
            "source_id": idx + 1,
            "parsed_content": parsed_json
        })

        print(f"Source {idx+1} Processed Successfully")

    except Exception as e:

        print(f"Error in Source {idx+1}: {e}")

# =====================================
# Save JSON
# =====================================
with open(
    "source_summaries.json",
    "w",
    encoding="utf-8"
) as f:

    json.dump(
        Final_summaries,
        f,
        # FILE OBJECTS ARE SAVED IN A HUMAN-READABLE FORMAT
        indent=4,
        ensure_ascii=False
    )

print("\nSaved Successfully!")
print("File: source_summaries.json")

````

````python
import os
import json
from dotenv import load_dotenv
from langchain_groq import ChatGroq

load_dotenv()

api_key = os.getenv("KEY")

if not api_key:
    raise ValueError("KEY not found.")

client = ChatGroq(
    groq_api_key=api_key,
    model_name="llama-3.3-70b-versatile",
    temperature=0.3,
)

# Read the summaries from the JSON file
with open("source_summaries.json", "r", encoding="utf-8") as f:
    summaries = json.load(f)


combined_summary = ""

for source in summaries:

    summary = source["parsed_content"]["summary"]

    combined_summary += summary + "\n\n"

print("="*80)
print("TOTAL SUMMARY LENGTH :", len(combined_summary))
print("="*80)

prompt = f"""
You are a journalist and professional news editor working for a globally recognized news agency.
Your responsibility is to transform the provided summary into a publication-ready news article that follows professional journalistic standards, maintains factual accuracy, and delivers information in a clear, structured, and engaging manner.

It should follow all the instructions mentioned below:
•	Carefully analyze the entire summary before writing.
•	Generate a clear, informative, and engaging main headline that accurately represents the central event.
•	Organize the article into 3 to 4 well-developed sections, each with a meaningful subheading and a detailed paragraph.
•	Ensure the article follows a logical flow, covering the event, background, key developments, overall significance and should answer in such a that

1.	What happened?
2.	Where did it happen?
3.	When did it happen?
4.	Who was involved?
5.	Why is it important?

•	Preserve every important factual detail from the summary, including names of people, organizations, locations, dates, statistics, events, and timelines.
•	Do not invent, assume, exaggerate, or omit factual information.
•	Maintain a formal, objective, and neutral journalistic tone throughout the article.
•	Follow The Guardian-style editorial principles by ensuring clarity, accuracy, balanced reporting, readability, and responsible journalism.
•	Avoid sensationalism, clickbait, biased opinions, offensive language, slang, informal expressions, and unsupported claims.
•	Produce a coherent, informative article of approximately 500 words that is suitable for publication on a professional news platform.

Output Format

Main Headline
Introduction
para
Subheading 1
para
Subheading 2
para
Subheading 3
para
"""


print("\nGenerating Final Article...\n")

response = client.invoke(prompt)

article = response.content

with open(
    "final_article.md",
    "w",
    encoding="utf-8"
) as f:

    f.write(article)

print("Final Article Saved Successfully!")
print("File : final_article.md")

print("\n")
print(article)
````
